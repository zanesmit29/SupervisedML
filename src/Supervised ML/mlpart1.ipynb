### Machine Learning Specialization by DeepLearning.AI Part 1


What is machine learning?
* The field of study that gives computers the ability to learn without explicitly being hardcoded

Types of Machine learning  

*  Supervised Learning
    * Algorithms that learn with input to output mappings (X and Y)
    * Inputs have labels
    * The algorithm 'sees' these correct pairs and can therefore learn from them 
    * Then it tries to predict the output for a new input  
    
Examples are spam filters and recommender systems, self driving cars etc. **Regression** and **Classification** are the two most widely known.  
  
Regression can be used in case of predicting the price of a house given a set of attributes. It is mainly used to predict numbers.    

Classification can be used for Breast cancer detection for example. Your dataset can have tumors of various sized and then diagnosis of "1" or "0". This is different because we are trying to produce a small number of outputs for a given input e.g. "1" or "0".

Because there are just two possible "categories" this is a classification problem.

Even if it was three categories, it would still be classification. **In general it can be numeric or non-numeric categories**.

Classification tries to find a boundary in the plotting to be able to "classify" the category.

*  Unsupervised Learning
    * Given data does not have a output (y) for given inputs (x)
    * Therefore no labels are given
    * It uses the data to find some pattern or structures in the data and group similar data points together
    * Typically it would cluster the data - Called a **clustering** algorithm

Google news uses these types of algorithms to group news segments. It would for example find articles with similar words and then clusters these news articles. It finds it all by itself.

Clustering is also used to identify different groups of people, for example the DeepLearning.AI team used this to try and figure out what are the different types of motivation for doing courses.

**Anomaly detection** is another unsupervised learning algorithm that is used to find unusual data points.
It is used in for example fraud detection in the financial systems.

**Dimensionality reduction** lets you take a big data set and compress is to a small data set with losing as little information as possible.





### Linear Regression model

A good example is predicting the price of a house using the size.  

Data will have the sizes and also the right answers (prices) for each given house size.  

A Linear regression gets the 'best fit' line for the data and then uses that line to predict new prices for a given size of house.

**Terminology**:

* Data set used to train the model is used the 'training set'
* Standard notation for input is lower case x and is called a 'feature'
* Output variable, or target variable is notated as lower case y
* m = number of training examples
* ($x$,$y$) = single training example
* ($x^i$,$y^i$) is used to notate the i'th training example with I referring to a specific row in the table  

$f(x)= wx + b$ or $f(x)$ this represents the linear function.  

**Univariate** Linear regression, is Linear regression that has one variable as an input.









### Cost Functions  

It tells you how well the model is doing, so you can improve the model.

Within $f(x) = wx + b$, $w$ and $b$ us called the weights or parameters and these will be adjusted during the training.

With Linear regression you want to get the 'best fit' line.

The **cost function** takes $\frac{1}{2m}\sum_{i=1}^{m} x_i$($\hat{y} - y$)$^2$ which is called the error. This is done for the sum of all errors, from $i = 1$ to $m$. This is then divided by $m$ to get the average error. This is called the **squared error cost function**.

It is written as $J(w,b)$ = $\frac{1}{2m}\sum_{i=1}^{m} $($f_{w,b}(x^i) - y^i$)$^2$

Eventually you want to find values of $w$ and $b$ that makes the cost function small.

While $f(x)$ is a linear function the plot of $J(w)$ where $b=0$ will be a parabola as described below:

```python
import matplotlib.pyplot as plt
import numpy as np

# Create 100 x-values evenly spaced between -2 and 2
x = np.linspace(-2, 2, 100)
# Compute corresponding y-values for the parabola y = x^2
y = x ** 2

# Set up the plot
plt.plot(x, y)
plt.title("Graph of y = xÂ²")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

Cost function $J$ when $b$ and $w$ is not zero, is a contour plot (3d-plot).  



### Training a model with *Gradient Descent*

**The gradient descent algorithm can be used to find the the values for $w$ and $b$ for which $J$ will be the smallest.**

$J$ is the cost function for Linear regression, but the Gradient Descent can also be used on the cost functions of any algorithms.  

Therefore the function is:   $min_{w_1,...,w_n,b}$ for $J(w_1,w_2,...,w_n,b)$  

Basically keep changing $w$ and $b$ until we find where J is the minimum. Depending on the function $J$ there may be multiple minimums.  

Depending on your starting point, the Gradient Descent algorithm identifies different local minima. Therefore, you can end up with different values.



**The gradient descent algorithm**: $w = w - \alpha$ $\frac{\partial }{\partial w}J(w,b)$
* $\alpha$ is the *Learning rate* and it controls how big a step downhill you take
* $\frac{\partial }{\partial w}J(w,b)$ is the *derivative* of the cost function $J$ and this tells you in which direction you want to take your steps. 
* Combined both tell you the size and direction of your steps

You will also have to update $b$ as it is also a variable: $b = b - \alpha$ $\frac{\partial }{\partial b}J(w,b)$


One more detail to do gradient descent: You want to simultaneously update $w$ and $b$ to the new values e.g. in code calculate 'temp_w' and 'temp_b' before assigning it to w and b as you will use both values to update the variables simultaneously.  

### The (partial) derivatives of *Gradient Descent*

The function $J$ where $b$ and $w$ is set to zero depending on the partial derivative.  

e.g. $w = w - \alpha$ $\frac{\partial }{\partial w}J(w)$ where $b=0$  

Therefore you will be calculating the derivative of a hyperbola, which is calculating the slope if the line touching the hyperbola at that point of the function. Therefore you will end up with something like:

$w = w - \alpha (positive\:number)$ where the calculated slope is positive, and
$w = w - \alpha (negative\:number)$ where the calculated slope is negative

### The Learning Rate

For a $\alpha$ too small the 'steps' is miniscule and the gradient descent takes a very long time. It needs many 'steps' to get there.

For a $\alpha$ too large It can be that you overshoot the minimum every time you try to find it. Which can actually have the opposite effect, you may actually go further away from the minimum and may *diverge* or fail to *converge*.

**What if you are already at a minimum?** If you draw a slope that touches the line it would be equal to 0 and therefore:  

$\frac{\partial}{\partial w}J(w) = 0$ and therefore the Gradient descent update will be updated as follows: $w = w - \alpha (0)$ and therefore $w = w$


As we get closer to a minimum, the updates will get smaller and smaller for a fixed learning rate.

Squared error cost function is that it will always converge to the local minimum. Therefore, this function is called a **convex** function.
